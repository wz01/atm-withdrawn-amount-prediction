```python
lmt_hidden = 1125000
```

```python
i =0
for index, row in df.iterrows():
    i = i + row['Total amount Withdrawn']
    if i >= lmt_hidden:
        ifor_val = 1
        df.at[index,'when need to refill'] = ifor_val
        i = 0;
    print(i)
```

```python
i =0
for index, row in df.iterrows():
    i = i + row['Total amount Withdrawn']
    df.at[index,'Accumulate Amount'] = i
    if i >= lmt_hidden:
        i = 0;
    print(i)
```
